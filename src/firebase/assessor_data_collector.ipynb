```python
import os
from os import listdir
from os.path import isfile, join
cwd = os.getcwd()
all_dirs = [dir for dir in listdir(cwd) if not isfile(join(cwd,dir)) and not dir.startswith(".")]
```

```python
import pandas as pd 
```

```python
users = []
for user_dir in all_dirs:
    user = {}
    user_info_filepath = join(join(cwd,user_dir),"user_info.txt")
    assessed_info_filepath = join(join(cwd,user_dir),"assessed_queries.txt")
    with open(user_info_filepath,"r") as infile:
        for line in infile:
            [key,value] = line.rstrip().split()
            key = key.rstrip(":")
            user[key] = value
        users.append(user)
    assessed_queries = []
    with open(assessed_info_filepath,"r") as infile:
        for line in infile:
            assessed_queries.append(line.rstrip())
    user["assessed_queries"] = assessed_queries
```

```python
df = pd.DataFrame(users,columns=["user_id","email","consent","contact","english","progress","assessed_queries","batches","unassessed_queries"]) 
```

```python
df['progress'] = df['progress'].astype(int)
df['english'] = df['english'].astype(int)
```

```python
df.sort_values(by=['progress'])
```

```python
output_df = df.drop(['batches', 'unassessed_queries'], axis=1)
```
