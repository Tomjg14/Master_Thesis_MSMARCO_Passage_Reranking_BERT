## Import Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.patches import ConnectionPatch
import numpy as np
```

## Results

```python
# precision
bm25_p5 = {"ms marco": 0.135, "t=2": 0.833, "t=3": 0.651}
bert_p5 = {"ms marco": 0.165, "t=2": 0.679, "t=3": 0.600}
bm25_p10 = {"ms marco": 0.081, "t=2": 0.700, "t=3": 0.526}
bert_p10 = {"ms marco": 0.093, "t=2": 0.507, "t=3": 0.428}

# map
bm25_map5 = {"ms marco": 0.436, "t=2": 0.947, "t=3": 0.877}
bert_map5 = {"ms marco": 0.672, "t=2": 0.848, "t=3": 0.819}
bm25_map10 = {"ms marco": 0.456, "t=2": 0.921, "t=3": 0.807}
bert_map10 = {"ms marco": 0.686, "t=2": 0.800, "t=3": 0.777}

# mrr
bm25_mrr5 = {"ms marco": 0.436, "t=2": 0.977, "t=3": 0.907}
bert_mrr5 = {"ms marco": 0.672, "t=2": 0.914, "t=3": 0.884}
bm25_mrr10 = {"ms marco": 0.456, "t=2": 0.977, "t=3": 0.907}
bert_mrr10 = {"ms marco": 0.686, "t=2": 0.914, "t=3": 0.884}

# mfr
bm25_mfr5 = {"ms marco": 3.44, "t=2": 1.05, "t=3": 1.26}
bert_mfr5 = {"ms marco": 2.21, "t=2": 1.28, "t=3": 1.37}
bm25_mfr10 = {"ms marco": 4.53, "t=2": 1.05, "t=3": 1.37}
bert_mfr10 = {"ms marco": 2.67, "t=2": 1.28, "t=3": 1.49}
```

```python
keys_ordered = ["ms marco", "t=2", "t=3"]

models = ["bm25", "bert"]

results = {}

results["bm25"] = {}
results["bert"] = {}

results["bm25"]["p@5"] = [bm25_p5[key] for key in keys_ordered]
results["bert"]["p@5"] = [bert_p5[key] for key in keys_ordered]

results["bm25"]["p@10"] = [bm25_p10[key] for key in keys_ordered]
results["bert"]["p@10"] = [bert_p10[key] for key in keys_ordered]

results["bm25"]["map@5"] = [bm25_map5[key] for key in keys_ordered]
results["bert"]["map@5"] = [bert_map5[key] for key in keys_ordered]

results["bm25"]["map@10"] = [bm25_map10[key] for key in keys_ordered]
results["bert"]["map@10"] = [bert_map10[key] for key in keys_ordered]

results["bm25"]["mrr@5"] = [bm25_mrr5[key] for key in keys_ordered]
results["bert"]["mrr@5"] = [bert_mrr5[key] for key in keys_ordered]

results["bm25"]["mrr@10"] = [bm25_mrr10[key] for key in keys_ordered]
results["bert"]["mrr@10"] = [bert_mrr10[key] for key in keys_ordered]

results["bm25"]["mfr@5"] = [bm25_mfr5[key] for key in keys_ordered]
results["bert"]["mfr@5"] = [bert_mfr5[key] for key in keys_ordered]

results["bm25"]["mfr@10"] = [bm25_mfr10[key] for key in keys_ordered]
results["bert"]["mfr@10"] = [bert_mfr10[key] for key in keys_ordered]
```

## Plot Results

### Plot Precision and MAP

```python
def fill_subplot(ax,metric,ylim):
    for model in models:
        ax.plot(keys_ordered,results[model][metric],label=model,marker='o',markerfacecolor='white',markersize=8)
    ax.set_ylim(ylim)
    ax.set_ylabel(metric)
    ax.yaxis.grid(linestyle='--',alpha=0.5)
```

```python
fig, axs = plt.subplots(2,2,figsize=(10,10))
fill_subplot(axs[0,0],"p@5",[0,1])
fill_subplot(axs[0,1],"p@10",[0,1])
fill_subplot(axs[1,0],"map@5",[0,1])
fill_subplot(axs[1,1],"map@10",[0,1])
handles, labels = axs[0,0].get_legend_handles_labels()
fig.legend(handles, labels, loc="lower center", fontsize='large', ncol=2)
fig.savefig("results_precision_and_map.png", bbox_inches='tight')
```

### Plot MRR and MFR

```python
fig, axs = plt.subplots(2,2,figsize=(10,10))
fill_subplot(axs[0,0],"mrr@5",[0,1])
fill_subplot(axs[0,1],"mrr@10",[0,1])
fill_subplot(axs[1,0],"mfr@5",[0,10])
fill_subplot(axs[1,1],"mfr@10",[0,10])
handles, labels = axs[0,0].get_legend_handles_labels()
fig.legend(handles, labels, loc="lower center", fontsize='large', ncol=2)
fig.savefig("results_mrr_and_mfr.png", bbox_inches='tight')
```
